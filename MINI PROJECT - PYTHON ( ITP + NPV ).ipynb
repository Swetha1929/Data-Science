### 1. To read the rounds2 and companies files, you would typically use Pandas read_csv function.

```python
import pandas as pd

# Read the rounds2 file
rounds2 = pd.read_csv("rounds2.csv", encoding="ISO-8859-1")

# Read the companies file
companies = pd.read_csv("companies.txt", sep="\t", encoding="ISO-8859-1")
```

```python
companies
```

### 2. To find the number of unique companies in rounds2, you can use the nunique() function.


```python
unique_rounds2_companies = rounds2['company_permalink'].nunique()
```

```python
unique_rounds2_companies
```

### 3. Similarly, for unique companies in the companies file:


```python
unique_companies = companies['permalink'].nunique()
```

### 4. To convert the permalink and company_permalink columns to uppercase:


```python
companies['permalink'] = companies['permalink'].str.upper()
rounds2['company_permalink'] = rounds2['company_permalink'].str.upper()
```

```python
companies['permalink'] 
```

```python
rounds2['company_permalink']
```

### 5. To check if there are any companies in rounds2 not present in companies:


```python
missing_companies = rounds2[~rounds2['company_permalink'].isin(companies['permalink'])]
```

```python
missing_companies
```

### 6. Merging the two dataframes:


```python
master_frame = pd.merge(rounds2, companies, how='left', left_on='company_permalink', right_on='permalink')

# And to get the number of observations:

observations = master_frame.shape[0]
```

```python
master_frame
```

```python
observations
```

### 7. Dropping the redundant column:



```python
master_frame.drop(columns=['company_permalink'], inplace=True)
```

### 8. To find the percentage of missing values and drop unnecessary columns:


```python
missing_percentage = master_frame.isnull().mean() * 100
master_frame.dropna(axis=1, how='all', inplace=True)
```

```python
missing_percentage
```

### 9. Creating a pivot table:


```python
pivot_table = master_frame.pivot_table(values='raised_amount_usd', index='funding_round_type', aggfunc=['mean', 'median'])
```

```python
pivot_table
```

### 10. Filtering master_frame based on raised_amount_usd:


```python
filtered_master_frame = master_frame[(master_frame['raised_amount_usd'] >= 4000000) & (master_frame['raised_amount_usd'] <= 15000000)]
```

```python
filtered_master_frame
```

### 11. Identifying top 9 countries in terms of highest investment in 'venture' fund type:


```python
top_9_countries = master_frame[master_frame['funding_round_type'] == 'venture'].groupby('country_code')['raised_amount_usd'].sum().nlargest(9)
```

```python
top_9_countries
```

### 12. Creating the 'main_category' column:

```python
# Define master_dataframe
master_dataframe = pd.merge(rounds2, companies, how='left', left_on='company_permalink', right_on='permalink')

# Define function to extract main category
def extract_main_category(category_list):
    if isinstance(category_list, str):
        return category_list.split('|')[0]
    else:
        return category_list

# Apply function to category_list column
master_dataframe['main_category'] = master_dataframe['category_list'].apply(extract_main_category)
```

```python
master_dataframe['main_category']
```

### 13. Converting 'raised_amount_usd' into categorical column:


```python
def recode_amount(amount):
    if amount < 5000000:
        return 'TypeA'
    elif amount < 7000000:
        return 'TypeB'
    else:
        return 'TypeC'

master_frame['raised_amount_category'] = master_frame['raised_amount_usd'].apply(recode_amount)
```
